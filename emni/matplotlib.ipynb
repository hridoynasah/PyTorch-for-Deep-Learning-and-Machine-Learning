**Subplot:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample 2x2 NumPy arrays with raw numbers
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

# Flatten arrays for linear plotting
flat1 = array1.flatten()  # [1, 2, 3, 4]
flat2 = array2.flatten()  # [5, 6, 7, 8]

# Create a 2x2 subplot grid
fig = plt.figure(figsize=(6, 6))

# Plot first array linearly
plt.subplot(2, 2, 1)  # (rows, cols, index)
plt.plot(flat1, marker='o')  # Linear plot with markers
plt.title("Array 1")
plt.grid(True)

# Plot second array linearly
plt.subplot(2, 2, 2)
plt.plot(flat2, marker='o')
plt.title("Array 2")
plt.grid(True)

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create three 2x2 NumPy arrays with raw numbers
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])
array3 = np.array([[9, 10], [11, 12]])

# Flatten arrays for linear plotting
flat1 = array1.flatten()  # [1, 2, 3, 4]
flat2 = array2.flatten()  # [5, 6, 7, 8]
flat3 = array3.flatten()  # [9, 10, 11, 12]

# Create a 2x2 subplot grid
fig = plt.figure(figsize=(6, 6))

# Plot first array linearly
plt.subplot(2, 2, 1)  # (rows, cols, index)
plt.plot(flat1, marker='o')
plt.title("Array 1")
plt.grid(True)

# Plot second array linearly
plt.subplot(2, 2, 2)
plt.plot(flat2, marker='o')
plt.title("Array 2")
plt.grid(True)

# Plot third array linearly
plt.subplot(2, 2, 3)
plt.plot(flat3, marker='o')
plt.title("Array 3")
plt.grid(True)

plt.tight_layout()  # Adjust spacing
plt.show()
```
