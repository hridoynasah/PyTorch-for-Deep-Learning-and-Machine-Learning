- Data (Prepareing and loading)

```python
import torch
from torch import nn
import matplotlib.pyplot as plt

torch.__version__
```

```python
# Create know parameters
weight = 0.7 
bias = 0.3

# Create data

start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step)
y = weight * X + bias
X.ndim, y.ndim
```

```python
X[:3], y[:3]
```

```python
X = torch.unsqueeze(X,dim = 1)
y = torch.unsqueeze(y,dim = 1)
X[:3], y[:3]
```

- **Split Data**

```python
# Create a train/test split
train_split = int(len(X) * .8)
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
len(y_train), len(X_test), len(y_test)
```

```python
def plot_predictions(train_data = X_train,
                     train_labels= y_train,
                     test_data = X_test,
                     test_labels = y_test,
                     predictions = None):
    plt.figure(figsize=(10, 7))
    
    # Plot training data in blue 
    plt.scatter(train_data, train_labels, c ='b', s = 50, label = 'Training data')
    # s is dot size
    plt.scatter(test_data, test_labels, c = 'g', s = 50, marker='x', label = 'Testing data')
    if predictions is not None:
        plt.scatter(test_data, predictions, c = 'r', s= 50, label = 'Predictions')
    
    # Show the legend
    plt.legend(prop={"size": 20})  
```

```python
plot_predictions()
```
